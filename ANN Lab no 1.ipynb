# Loops

1. For Loop

```python
for i in range(3):
  print(i)
```

2. While Loop

```python
count = 1
while count <=3:
    print(count)
    count +=1
```

3. Nested Loop

```python
for i in range(3):
  for j in range(4):
    print(i,j)
```

4. Loop With Break

```python
for i in range(5):
   if i == 3:
      break
   print(i)
```

5. Loop with Continue

```python
for i in range(5):
  if i == 3:
    continue
  print(i)
```

## **Functions**

1. Function Without Parameters

```python
def weight():
    print("My weight is 65kg")
weight()
```

2. Function With Parameters

```python
def add(a, b):
    print(a + b)
add(2, 7)
```

3. Function With Return Value

```python
def square(n):
    return n * n
result = square(3)
print(result)
```

4. Default Parameter Function

```python
def greet(name="Student"):
    print("Hello", name)
greet()
greet("Masood")
```

5. Keyword Arguments

```python
def student(name, age):
    print(name, age)
student(age=23, name="Masood")
```

6. Lambda Function

```python
square = lambda x: x*x
print(square(7))
```

## **Lists**

1. Adding a New Element at End

```python
numbers = [1, 2, 3, 4]
numbers.append(5)
print(numbers)
```

2. Adding More than 1 Element

```python
numbers = [1, 2, 3, 4]
numbers.extend([6,7])
print(numbers)
```

3. Inserting A Number Between a List

```python
numbers = [1, 2, 3, 4]
numbers.insert(1, 10)
print(numbers)
```

4. Removing an Element

```python
numbers = [1, 2, 3, 4]
numbers.remove(3)
print(numbers)
```

5. Removing an Element using Pop

```python
numbers = [1, 2, 3, 4]
numbers.pop()
print(numbers)
```

6. Finding The Index of a Number

```python
numbers = [1, 2, 3, 4]
index = numbers.index(1)
print(f"The element at index {index} is: {numbers[index]}")
```

7. Couting the Repeated elements

```python
numbers = [1, 2, 1, 4]
numbers.count(1)
print(numbers.count(1))
```

8. Sorting the List in Ascending Order

```python
numbers = [2, 1, 4, 5, 3]
numbers.sort()
print(numbers)
```

9. Sorting the List in Descending Order

```python
numbers = [1, 2, 3, 4]
numbers.reverse()
print(numbers)
```

10. Coping a List

```python
numbers = [1, 2, 3, 4]
new_list = numbers.copy()
print(new_list)
```

11. Deleting all Elements in the List

```python
numbers = [1, 2, 3, 4]
numbers.clear()
print(numbers)
```

12. Length of a List

```python
numbers = [1, 2, 3, 4]
len(numbers)
```

# **Dictionaries**

1. Creating a Dictionary

```python
teacher = {"name": "Mr.Saeed", "subject": "ANN"}
print(teacher)
```

2. Accessing Values

```python
print(teacher["name"])
```

3. Adding/Updating

```python
teacher["department"] = "Ai"
teacher["subject"] = "Machine Learning"
```

4. List to Dictionary

```python
keys = ["name", "age"]
values = ["Masood", 21]
new_dict = dict(zip(keys, values))
print(new_dict)
```

5. Dictionary to List

```python
dict_items = list(teacher.items())
print(dict_items)
```

6. Dictionary with Tuples

```python
locations = {(5,7): "Point A", (6,9): "Point B"}
print(locations)
```

## **2. Vectorized Operations**

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)
print(a * b)
print(a ** 3)
print(b ** 2)
```

## **3. Matrix Operations**

1. Matrix Multiplication

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
result = np.dot(A, B)
print(result)
```

2. Broadcasting

```python
a = np.array([[1, 2, 3],[4, 5, 6]])
b = np.array([10, 20, 30])
print(a + b)
```
